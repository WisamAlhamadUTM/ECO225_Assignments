## Part 1: Python Data Structures

### 1. Input and Inspect the Data

```python
#Input the data manually to Python (no Pandas).
#Store rows as a list of lists.
data = [["country", "year", "gdp", "unemployment", "inflation"],
    ["Canada", 2024, 2.1, 5.4, 3.9],
    ["USA", 2024, 2.6, 3.8, 3.2],
    ["UK", 2024, 1.4, 4.2, 4.7],
    ["Germany", 2024, 0.8, 3.1, 2.8],
    ["Japan", 2024, 1.0, 2.6, 2.3]]

print(data)

#Print the first and last rows
print(data[0])
print(data[-1])
# Check the data type of each row
for row in data:
    print(type(row))

#Save the list of lists as a CSV to your pythonanywhere (import the csv package)
import csv

with open("economic_data.csv", "w", newline="") as f:
    writer = csv.writer(f)
    writer.writerows(data)
```

### 2. Convert Data into Multiple Data Structures

```python
#Convert rows into list of dictionaries.

list_of_dicts = []

for row in data[1:]:
    d = {}
    d["country"] = row[0]
    d["year"] = row[1]
    d["gdp"] = row[2]
    d["unemployment"] = row[3]
    d["inflation"] = row[4]
    list_of_dicts.append(d)

print(list_of_dicts)
print("\n")

# Create a set of unique countries
unique_countries = {row[0] for row in data[1:]}
print(unique_countries)
print("\n")

# Convert GDP values to a tuple
gdp_tuple = tuple(row[2] for row in data[1:])
print(gdp_tuple)
print("\n")

# Create dictionary mapping Country -> GDP
gdp_c_map = {}
for row in data[1:]:
    gdp_c_map[row[0]] = row[2]

print(gdp_c_map)
```

### 3. Python Control Structures

```python
# Conditionals: Categorize GDP as Low / Moderate / High Growth
for row in data[1:]:
    if row[2] > 2.5:
        print(row[0], "High Growth")
    elif row[2] >= 1.5:
        print(row[0], "Moderate Growth")
    else:
        print(row[0], "Low Growth")
```

```python
# For Loops: Identify countries with unemployment < 4
print("Low Unemployment Countries:")
for row in data[1:]:
    if row[3] < 4:
        print(row[0])
```

```python
# While Loops: Multiply all GDP values by 100
i = 1
while i < len(data):
    data[i][2] = data[i][2] * 100
    i += 1

print(data)
```

```python
# Iterators: Create an iterator for GDP tuple
gdp_iter = iter(gdp_tuple)

print(next(gdp_iter))
print(next(gdp_iter)) 

for gdp in gdp_iter:
    print(gdp)
```

### 4. String Manipulation

```python
#Using the string “Economic Indicators"

s = "Economic Indicators"

#Slice the 3rd to 6th characters.
print(s[2:6])
#Slice every 3rd character
print(s[::3])
#Slice the last 10 characters.
print(s[-10:])
#Extract “Economic” using slicing.
print(s[:9])
#Convert this to all capitals using string methods.
print(s.capitalize())
```

### 5. Functions and OOP (5 marks)

```python
#Function:
#Define a function country gdp()
#Make the function return GDP for a country when the country name is given
#Call the function for all the countries in the economic data given

def country_gdp(country):
    return gdp_c_map[country]

print(country_gdp("USA"))
print(country_gdp("UK"))
print(country_gdp("Japan"))
print(country_gdp("Canada"))
print(country_gdp("Germany"))
      
```
